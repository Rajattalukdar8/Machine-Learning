**Hierachial  Clustering**

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dataset
data=pd.read_csv("/content/Mall_Customers.csv")
x=data.iloc[:,[3,4]].values
```

using dendogram to find optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Eucledian distances')
plt.show()
```

Training the Hierachial model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(x)
```

```python
y_hc
```

Visualizing the clusters

```python
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,c='magenta',label='cluster 1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100,c='green',label='cluster 2')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=100,c='cyan',label='cluster 3')
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],s=100,c='yellow',label='cluster 4')
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],s=100,c='orange',label='cluster 5')

plt.title('Clusters of customers')
plt.xlabel('Anual income')
plt.ylabel('spending score')
plt.legend()
plt.show()

```
